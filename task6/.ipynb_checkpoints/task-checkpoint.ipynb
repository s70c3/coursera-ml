```python
import pandas as pd
import numpy as np
import pylab
import math

from skimage.io import imread
from skimage import img_as_float
from sklearn.cluster import KMeans

%matplotlib inline
```

```python
def psnr(img1, img2):
    mse = np.average(np.mean( (img1 - img2) ** 2))
    if mse == 0:
        return 100
    PIXEL_MAX = 1
    return 20 * math.log10(PIXEL_MAX / np.sqrt(mse))
```

```python
img = imread('parrots.jpg')
img = img_as_float(img)
print(img.shape)
matrix = img.reshape(-1,3)
pylab.imshow(img)
```

```python
print(matrix)
```

```python
df = pd.DataFrame(matrix)
```

```python
def new_image(img, y, clusters, ttype='mean'):
    df['cluster'] = y
    if ttype=='mean':
        colors = df.groupby('cluster').mean()
    elif ttype=='median':
        colors = df.groupby('cluster').median()
    df['r']=0
    df['g']=0
    df['b']=0
    for i in range(clusters):
        df.loc[df['cluster']==i, 'r']=colors.loc[i][0]
        df.loc[df['cluster']==i, 'g']=colors.loc[i][1]
        df.loc[df['cluster']==i, 'b']=colors.loc[i][2]
    new_img = df.loc[:,'r':]
    return new_img
```

```python
for i in range(1,20):
    cls = KMeans(random_state=241, init='k-means++', n_clusters=i)
    y = cls.fit_predict(matrix)
    
    mean_img = new_image(matrix, y, i, 'mean')
    median_img = new_image(matrix, y, i, 'median')

    
    print(str(i), 'mean', str(psnr(mean_img, matrix)))
    print(str(i), 'median', str(psnr(median_img, matrix)))
    n=np.array(mean_img).reshape(474, 713, 3)
    pylab.imshow(mean_img)
    n=np.array(median_img).reshape(474, 713, 3)
    pylab.imshow(median_img)
```

```python
cls = KMeans(random_state=241, init='k-means++', n_clusters=11)
y = cls.fit_predict(matrix)
mean_img = new_image(matrix, y, 11, 'mean')
n=np.array(mean_img).reshape(474, 713, 3)
pylab.imshow(n)
```
